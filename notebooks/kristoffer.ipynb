```python
import torch
import wandb
import shutil
import os
from jormungandr.config.configuration import load_config
from jormungandr.jormungandr import Jormungandr
```

```python
artifact_name = "sverrenystad-ntnu/Fafnir/mamba_giouloss_20260423_172048:v40"
```

```python
api = wandb.Api()
artifact = api.artifact(artifact_name, type="model")
artifact_dir = artifact.download()
checkpoint_path = os.path.join(artifact_dir, os.listdir(artifact_dir)[0])
state_dict = torch.load(checkpoint_path)

```

```python
new_state_dict = {}

for k, v in state_dict.items():
    if k.startswith("encoder."):
        new_key = k.replace("encoder.", "spatial_encoder.")
        new_state_dict[new_key] = v

    elif k.startswith("decoder."):
        new_state_dict[k] = v

    elif k.startswith("output_head."):
        new_state_dict[k] = v
```

```python
new_state_dict.keys()
```

```python
config = load_config("experiment_5_2.yaml")
```

```python
jormungander = Jormungandr(device="cuda", config=config.model)
```

```python
jormungander.load_state_dict(new_state_dict, strict=False)
```
